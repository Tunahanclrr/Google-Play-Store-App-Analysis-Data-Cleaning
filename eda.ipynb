### Standardizing App Size Values in the Dataset
The Size column in the dataset represented the storage space required by each application. However, the values were stored in different formats:

100M, 200M → Megabytes

200k → Kilobytes

To standardize the data, I performed the following steps:

Created a new column named Size (KB).

For megabyte values, I removed the "M" character and added "000" at the end to convert them into kilobytes.

Example: 100M → 100000 KB

For kilobyte values, I removed the "k" character and kept the number as it is.

Example: 200k → 200 KB

Dropped the original Size column from the dataset.

 Result: All app sizes are now standardized and expressed in kilobytes (KB).

```python
df["Size"].value_counts()
```

```python
df["Size"]=df["Size"].str.replace("M","000")
df["Size"]=df["Size"].str.replace("k","")
```

### Data Cleaning: Cleaning "Installs", "Price", and "Date" Columns

In the dataset:  
- The `installs` column contained values like `10,200+`, `5,000+`, etc.  
- The `price` column contained values like `$1,99`, `$0`, etc.  
- The `date` column was stored as strings (e.g., `"January 1, 2020"`), which are not suitable for date operations.

Cleaning steps:  
1. Removed "+" signs and commas from `installs` to convert them into integers.  
2. Removed "$" signs and commas from `price` to convert them into floats.  
3. Converted the `date` column from string to datetime format for proper date handling.

Examples:

- `installs`  
  - Before: `10,200+`  
  - After: `10200`

- `price`  
  - Before: `$1,99`  
  - After: `1.99`

- `date`  
  - Before: `"January 1, 2020"`  
  - After: `2020-01-01`

### Data Cleaning and Analysis Process

This process covers the steps applied to transform raw data into analyzable and meaningful insights.

1. Android Version (Android Ver) Column Cleaning

The column containing the Android versions supported by applications was prepared for analysis.

Problem Identification:
The column contained irregular values such as "and up" and "Varies with device", which prevented proper analysis.

Solution Steps:

The "and up" string was removed using the replace() function.

"Varies with device" was replaced with NaN.

Android versions were not converted to float, in order to avoid losing detailed version information (e.g., 4.0.3). Instead, the values were preserved as strings for accurate analysis.

As a result, the most common Android versions were identified as 4.1, 4.0.3, and 4.4.

2. Application Ratings (Rating) Analysis

This step focused on examining the rating scores of applications to identify the highest-rated ones.

The data was grouped by Category and App.

Ratings were aggregated and sorted in descending order.

Applications with a perfect score of 5.0 were filtered.

In total, 271 applications achieved a 5.0 rating, with categories such as FAMILY, TOOLS, FINANCE, and PARENTING being prominent.

🎯 Conclusion

This analysis highlights how data cleaning and transformation (especially standardizing the Android Ver column without converting to float) combined with rating analysis allow valuable insights to be extracted from the dataset.
